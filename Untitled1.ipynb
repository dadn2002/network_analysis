```python
import networkx as nx
G = nx.DiGraph()
nodes_list = ['M', 'N', 'O', 'P', 'Q', 'R', 'S']
edges_list = [('M', 'S'), ('N', 'O'), ('P', 'R'), ('N', 'S'), ('O', 'M'), ('N', 'Q'), ('O', 'M'), ('P', 'P'), ('S', 'M'), ('O', 'N'),  ('S', 'M'), ('N', 'R'), ('P', 'M'), ('M', 'S')]
G.add_nodes_from(nodes_list)
G.add_edges_from(edges_list)

nx.draw_networkx(G)
print("Questão 6")
print("Exist path M to S?:", nx.has_path(G, 'M', 'S'), "\nShortest path:", nx.shortest_path(G, 'M', 'S'))
print("\nList of cicles of any order of G:\n", sorted(nx.simple_cycles(G, 4)),"\n So no cycle of order 4")
print("\nIs the graph connected?:\n", nx.is_semiconnected(G))
print("\nDegree of Nodes:\n",nx.degree(G), "\n Degree of N is", nx.degree(G, 'N'), " and Degree of R is", nx.degree(G, 'R'))
print("\nGraph to list of adjacency:\n", sorted(nx.generate_adjlist(G)))
print("\nGraph to matrix:\n", nx.to_numpy_array(G))
```
